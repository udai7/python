```python
n=5
spaces=2*(n-1)

for row in range(1,n+1):
    for column in range(1,row+1):
        print(column, end="")
    for column in range(1, spaces+1):
        print(" ",end="")
    for column in range(row,0,-1):
        print(column,end="")
    print()
    spaces -= 2
```
