```python
import requests
url = "https://en.wikipedia.org/wiki/Mobile_country_code"

headers = {
    "User-Agent": "Mozilla/5.0"
}

response = requests.get(url, headers=headers)
tables = pd.read_html(StringIO(response.text), match="Country", header=0)

df = tables[0]
print(df.head())
```
