```python
import sqlite3
```

```python
## Connect to a SQLite Database
connection=sqlite3.connect('example.db')
connection
```

```python
cursor=connection.cursor()
```

```python
##Create a Table
cursor.execute('''
Create Table if Not Exists employees(
    id Integer Primary Key,
    name Text Not Null,
    age Integer,
    department text
    )
''')

##Commit the changes
connection.commit()
```

```python
cursor.execute('''
Select * from employees
''')
```

```python
##Insert the data in a sqlite table
cursor.execute('''
Insert Into employees(name,age,department)
               values('Doiee',22,'Student')
''')

cursor.execute('''
Insert Into employees(name,age,department)
               values('Jini',21,'Student')
''')

cursor.execute('''
Insert Into employees(name,age,department)
               values('Bhai',1000,'Student')
''')

##Commit code
connection.commit()
```

```python
##Query data from the table

cursor.execute('select * from employees')
rows=cursor.fetchall()

##Print the rows

for row in rows:
    print(row)
```

```python
##update the data in table

cursor.execute('''
UPDATE employees
Set age = 34
where name="Doiee"
''')

connection.commit()
```

```python
##Query data from the table

cursor.execute('select * from employees')
rows=cursor.fetchall()

##Print the rows

for row in rows:
    print(row)
```

```python
##Delete Data from table
cursor.execute('''
Delete from employees
               where name="Doiee"
''')

connection.commit()
```

```python
##Query data from the table

cursor.execute('select * from employees')
rows=cursor.fetchall()

##Print the rows

for row in rows:
    print(row)
```

```python
##Working with Sales data
##Connect with a SQLite database

connection = sqlite3.connect('sales_data.db')
cursor=connection.cursor()

#Create a table for sales data

cursor.execute('''
CREATE TABLE IF NOT EXISTS sales(
    id INTEGER PRIMARY KEY,
    date TEXT NOT NULL,
    product TEXT NOT NULL,
    sales INTEGER,
    region TEXT
    )
''')

connection.commit()
```

```python
#INSERT data into sales table
sales_data=[
    ('1/10/10', 'Chocobar', 100, 'east'),
    ('5/10/10', 'Chocobuiscuit', 23, 'north'),
    ('10/10/10', 'Choco icecream', 335, 'south'),
    ('15/10/10', 'Chocolate', 345, 'West')
]

cursor.executemany('''
Insert into sales(date,product,sales,region)
                   values(?,?,?,?)
''',sales_data)

connection.commit()
```

```python
cursor.execute('SELECT * FROM sales')
rows= cursor.fetchall()

for row in rows:
    print(row)
```

```python
##close the connection

connection.close()
```
